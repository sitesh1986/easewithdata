```python
%sql
-- Merge Data To Update Existing Records And Insert New Records
MERGE INTO dev.bronze.employee e
USING dev.bronze.employee_updates u
ON e.emp_id = u.emp_id 
when MATCHED THEN 
UPDATE 
    set e.emp_salary = u.emp_salary
when not matched then
INSERT *
WHEN NOT MATCHED by SOURCE THEN
DELETE;
```

```python
%sql
ALTER TABLE dev.bronze.employee
ADD COLUMNS (
    isActive BOOLEAN
);
```

```python
%sql
-- Merge Data To Update Existing Records And Insert New Records
MERGE INTO dev.bronze.employee e
USING dev.bronze.employee_updates u
ON e.emp_id = u.emp_id 
when MATCHED THEN 
UPDATE 
    set e.emp_salary = u.emp_salary,
        e.isActive = true
when not matched then
INSERT 
(
    emp_id, 
    emp_name, 
    emp_salary, 
    emp_dept, 
    isActive
) 
VALUES 
(
    u.emp_id, 
    u.emp_name, 
    u.emp_salary, 
    u.emp_dept, 
    true
)
WHEN NOT MATCHED by SOURCE THEN
UPDATE 
SET e.isactive = false;
```
