```python
%sql
-- If Not Exist Clause
CREATE TABLE IF NOT EXISTS dev.bronze.Employee(
    emp_id int,
    emp_name STRING,
    emp_salary DOUBLE,
    emp_dept STRING
);
insert into table dev.bronze.employee values (1,'John',50000,'IT'),(2,'Jane',60000,'HR'),(3,'Bob',70000,'IT');
select * from dev.bronze.employee
```

```python
%sql
-- Create views and temprory views
-- Temprory views
create temporary view emp_tmp_vw
as
select * from dev.bronze.employee where emp_salary >= 55000;
```

```python
%sql
select * from emp_tmp_vw
```

```python
%sql
create or replace  view dev.bronze.emp_vw
as
select * from dev.bronze.employee where emp_salary >= 55000;
```

```python
%sql
select * from dev.bronze.emp_vw;
```

```python
%sql
-- CTAS FOR DELTA TABLE
create table dev.bronze.employee_cts
as
select * from dev.bronze.employee
;
```

```python
%sql
select * from dev.bronze.employee_cts
```

```python
%sql
-- DESCRIBE CTAS TABLE
describe extended dev.bronze.employee_cts 
```

```python
%sql
-- DEEP CLONE copy both metadata and data
CREATE table dev.bronze.employee_dc DEEP CLONE dev.bronze.employee;
```

```python
%sql
describe extended dev.bronze.employee_dc;
```

```python
%sql
describe history dev.bronze.employee_dc;
```

```python
%sql
-- SHALLOW CLONE it will only copy metadata

CREATE Table dev.bronze.employee_sc SHALLOW CLONE dev.bronze.employee;
```
